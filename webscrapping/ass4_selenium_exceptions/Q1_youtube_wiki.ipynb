# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
##### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
##### You need to find following details:
##### A) Rank
##### B) Name
##### C) Artist
##### D) Upload date
##### E) Views

```python
import selenium
from selenium import webdriver
from time import sleep
import time
import pandas as pd

import requests # importing Request
import re  #importing Regex
import time

from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r'C:\Users\harsh\Desktop\chromedriver.exe')
driver.maximize_window()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Name of the video
Name= []
name_ = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]')
for i in name_:
    Name.append(i.text)
Name
```

```python

artist =[]
try:
    art = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[3]')
    for i in art:
        artist.append(i.text)
        
except NoSuchElementException:
     artist.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    artist.append('No details available')
        
artist
```

```python
#Views 
views = []
try:
    v = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[4]')
    for i in v:
        views.append(i.text)
        
except NoSuchElementException:
     views.append('--')
        
views
```

```python
#scraping artist/uploader of the video

artist=[]

try:

    artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")

    for i in artists:

        artist.append(i.text)

except NoSuchElementException:#handling no such element exception

    artist.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    artist.append('No details available')
    
artist
```

```python
#Upload Date
upload_d = []
try:
    upload = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//tbody/tr/td[5]')
    for i in upload:
        upload_d.append(i.text)
        
except NoSuchElementException:#handling no such element exception

    upload_d.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    upload_d.append('No details available')
    
upload_d
```

```python
#rank
rank = []
try:
    r = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in r:
        rank.append(i.text.replace('.',''))
        
except NoSuchElementException:#handling no such element exception

    rank.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    rank.append('No details available')
    
rank[0:30]
```

```python
video = pd.DataFrame({'Name':Name,'artist':artist,'views':views,'upload_d':upload_d[0:30],'Rank':rank[0:30]})
video
```
