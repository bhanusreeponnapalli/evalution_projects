# 4.Scrape the details of State-wise GDP of India from statisticstime.com.
##### Url = http://statisticstimes.com/
##### You have to find following details:
##### A) Rank
##### B) State
##### C) GSDP
##### D) GSDP
##### E) Share
##### F) GDP($ billion)


```python
import selenium
from selenium import webdriver
from time import sleep
import time
import pandas as pd

import requests # importing Request
import re  #importing Regex
import time
from selenium.webdriver.support.ui import Select

from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r'C:\Users\harsh\Desktop\chromedriver.exe')
driver.maximize_window()
```

```python
driver.get(' http://statisticstimes.com/')
```

```python
economy = driver.find_element_by_xpath('//div[@class="navbar"]/div[2]')
economy.click()
select = driver.find_element_by_xpath('//div[@class="dropdown-content"]/a[3]')
select.click()
```

```python
#selecting the indian states
indian= driver.find_element_by_xpath('//div[@style="float:left;background-color:seashell;width:400px;height:800px;"]/ul/li[1]/a')
indian.click()
```

```python
#rank
rank = []
r1 = []
try:
    r = driver.find_elements_by_xpath('//td[@class="data1"]')
    for i in r:
        r1.append(i.text)
except NoSuchElementException:
          r1.append('No details available')

except StaleElementReferenceException:#handling Stale element exception
        r1.append('No details available')
        
rank = r1[:33]
rank
```

```python
#State
state = []
s=[]
try:
    st = driver.find_elements_by_xpath('//td[@class="name"]')
    for i in st:
        s.append(i.text)
        
except NoSuchElementException:
          s.append('No details available')

except StaleElementReferenceException:#handling Stale element exception
        s.append('No details available')
state = s[:33]
state
```

```python
#data 
data = []
try:
    d = driver.find_elements_by_xpath('//td[@class="data"]')
    for i in d:
        data.append(i.text)
        
except NoSuchElementException:
          data.append('No details available')

except StaleElementReferenceException:#handling Stale element exception
        data.append('No details available')
        
data[:165]

GSDP_curr_19_20 = data[0:165:5]
share = data[1:165:5]
GDP_billions = data[2:165:5]
GSDP_2011_prz_19_20 = data[3:165:5]
GSDP_2011_prz_18_19 = data[4:165:5]
data = pd.DataFrame({'GSDP(curr_19_20)':GSDP_curr_19_20,
                    'Share':share,
                    'GDP($billions)':GDP_billions,
                    'GSDP_2011(prz_19_20)':GSDP_2011_prz_19_20,
                    'GSDP_2011(prz_18_19)':GSDP_2011_prz_18_19})
   
data
```

```python
#GSDP_curr_18_19
GSDP_curr_18_19 = []
curr_18_19 = []
try:
    curr= driver.find_elements_by_xpath('//td[@class="data sorting_1"]')
    for i in curr:
        curr_18_19.append(i.text)

except NoSuchElementException:
        data.append('No details available')

except StaleElementReferenceException:#handling Stale element exception
        data.append('No details available')        

GSDP_curr_18_19 = curr_18_19[:33]
GSDP_curr_18_19
```

```python
GDP =pd.DataFrame({'Rank':rank,
                  'State':state,
                  'GSDP(19-20)':GSDP_curr_19_20,
                  'GSDP(18-19)':GSDP_curr_18_19,
                  'Share':share,
                  'GDP($billions)':GDP_billions,
                   'GSDP(19-20)':GSDP_2011_prz_19_20,
                  'GSDP(18-19)':GSDP_2011_prz_18_19,
                  })
GDP
```
